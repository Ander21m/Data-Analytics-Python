```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import os


dataset = load_dataset("csv",data_files=os.path.join(os.path.abspath(os.path.join(os.getcwd(),"..")),"data_jobs.csv"))
df = dataset['train'].to_pandas()

df["job_posted_date"]= pd.to_datetime(df.job_posted_date)
```

```python
df.pivot_table(values = "salary_year_avg",index="job_title_short",aggfunc="median")
```

```python
df.pivot_table(values = "salary_year_avg",index="job_country",aggfunc="median",columns="job_title_short")
```

```python
top_countries = df["job_country"].value_counts().head(6).index
df_top_coutries = df.pivot_table(values = "salary_year_avg",index="job_country",aggfunc="median",columns="job_title_short")


df_top_coutry_jobs = df_top_coutries.loc[top_countries]

job_titles = ["Data Analyst","Data Engineer","Data Scientist"]


df_top_coutry_jobs = df_top_coutry_jobs[job_titles]
df_top_coutry_jobs
```

```python
df_top_coutry_jobs.plot(kind= "bar")

plt.ylabel("Median Salary ($USD)")
plt.xlabel("")
plt.title("Median Salary By Country and Job Title")
plt.xticks(rotation = 45,ha = "right")
plt.show()
```
