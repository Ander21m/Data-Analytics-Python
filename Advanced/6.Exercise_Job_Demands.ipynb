```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import os


dataset = load_dataset("csv",data_files=os.path.join(os.path.abspath(os.path.join(os.getcwd(),"..")),"data_jobs.csv"))
df = dataset['train'].to_pandas()

df["job_posted_date"]= pd.to_datetime(df.job_posted_date)
```

```python
df_usa = df[df["job_country"] == "United States"].copy()
df_usa
```

```python
df_usa["job_posted_month"] = df_usa["job_posted_date"].dt.strftime("%B")
```

```python
df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index ="job_posted_month",columns="job_title_short",aggfunc="size")
```

```python
df_usa_pivot.reset_index(inplace= True)
```

```python
df_usa_pivot["month_no"] = pd.to_datetime(df_usa_pivot.job_posted_month,format= "%B").dt.month
df_usa_pivot = df_usa_pivot.sort_values(by = "month_no")
```

```python
df_usa_pivot.set_index("job_posted_month",inplace=True)
```

```python
df_usa_pivot
```

```python


top3Jobs = df_usa["job_title_short"].value_counts().head(3)
top3Jobs = top3Jobs.index.to_list()
```

```python
df_usa_pivot[top3Jobs].plot(kind= "line")
plt.ylabel("Job Count")
plt.xlabel("2023")
plt.title("Monthly job posting for top jobs in USA")
plt.legend()
plt.show()
```
